```python
import numpy as np
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski
```

```python
def lipinski(smiles, verbose=False):

    moldata= []
    for elem in smiles:
        mol=Chem.MolFromSmiles(elem) 
        moldata.append(mol)
       
    baseData= np.arange(1,1)
    i=0  
    for mol in moldata:        
       
        desc_MolWt = Descriptors.MolWt(mol)
        desc_MolLogP = Descriptors.MolLogP(mol)
        desc_NumHDonors = Lipinski.NumHDonors(mol)
        desc_NumHAcceptors = Lipinski.NumHAcceptors(mol)
           
        row = np.array([desc_MolWt,
                        desc_MolLogP,
                        desc_NumHDonors,
                        desc_NumHAcceptors])   
    
        if(i==0):
            baseData=row
        else:
            baseData=np.vstack([baseData, row])
        i=i+1      
    
    columnNames=["MW","LogP","NumHDonors","NumHAcceptors"]   
    descriptors = pd.DataFrame(data=baseData,columns=columnNames)
    
    return descriptors
     

df_lipinski = lipinski(df.canonical_smiles)
```

```python
import numpy as np
def pIC50(input):
    pIC50 = []
    
    for i in input['standard_value_norm']:
        molar = i*(10**-9) # Converts nM to M
        # Add check to avoid log10(0)
        if molar > 0:
            pIC50.append(-np.log10(molar))
        else:
            pIC50.append(float('nan'))  # Use NaN for zero values
        
    input['pIC50'] = pIC50
    
    # Check which columns actually exist before dropping
    columns_to_drop = []
    if 'standard_value' in input.columns:
        columns_to_drop.append('standard_value')
    if 'standard_value_norm' in input.columns:
        columns_to_drop.append('standard_value_norm')
    
    # Only drop if there are columns to drop
    if columns_to_drop:
        x = input.drop(columns=columns_to_drop)
    else:
        x = input
    
    return x
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski

# Assuming df_final has a column with SMILES strings
# If your SMILES column has a different name, replace 'canonical_smiles' with your column name
smiles_column = 'canonical_smiles'

# Function to calculate descriptors
def calculate_descriptors(smiles):
    # Convert SMILES to RDKit molecule object
    mol = Chem.MolFromSmiles(smiles)
    
    if mol is None:  # If SMILES couldn't be converted
        return [np.nan] * 10  # Return NaN for all descriptors
    
    # Calculate descriptors
    descriptors = {
        'MW': Descriptors.MolWt(mol),                      # Molecular weight
        'LogP': Descriptors.MolLogP(mol),                  # Octanol-water partition coefficient
        'NumHDonors': Lipinski.NumHDonors(mol),            # Number of H-bond donors
        'NumHAcceptors': Lipinski.NumHAcceptors(mol),      # Number of H-bond acceptors
        'NumRotatableBonds': Descriptors.NumRotatableBonds(mol),  # Number of rotatable bonds
        'NumAromaticRings': Descriptors.NumAromaticRings(mol),    # Number of aromatic rings
        'TPSA': Descriptors.TPSA(mol),                     # Topological polar surface area
        'NumHeavyAtoms': Descriptors.HeavyAtomCount(mol),  # Number of heavy atoms
        'NumAtoms': mol.GetNumAtoms(),                     # Total number of atoms
        'NumRings': Descriptors.RingCount(mol)             # Number of rings
    }
    
    return pd.Series(descriptors)

# Apply the function to each SMILES string in the dataframe
try:
    # Check if SMILES column exists
    if smiles_column in df_final.columns:
        # Calculate descriptors and add them as new columns
        descriptors_df = df_final[smiles_column].apply(calculate_descriptors)
        
        # Combine with original dataframe
        df_with_descriptors = pd.concat([df_final, descriptors_df], axis=1)
        
        # Display the first few rows to verify
        print(df_with_descriptors.head())
        
        # Save to CSV if needed
        df_with_descriptors.to_csv('molecules_with_descriptors.csv', index=False)
        
        print(f"Descriptors calculated for {len(df_with_descriptors)} molecules")
    else:
        print(f"Error: Column '{smiles_column}' not found in the dataframe")
        print("Available columns:", df_final.columns.tolist())
except Exception as e:
    print(f"Error calculating descriptors: {e}")
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski

# Assuming df_final has a column with SMILES strings
# If your SMILES column has a different name, replace 'canonical_smiles' with your column name
smiles_column = 'canonical_smiles'

# Create empty lists to store descriptor values
mols = []
mw = []
logp = []
num_h_donors = []
num_h_acceptors = []
num_rotatable_bonds = []
num_aromatic_rings = []
tpsa = []
num_heavy_atoms = []
qed = []  # Drug-likeness score

# Calculate descriptors for each SMILES
for smiles in df_final[smiles_column]:
    mol = Chem.MolFromSmiles(smiles)
    
    if mol is None:  # If SMILES couldn't be converted
        mols.append(None)
        mw.append(np.nan)
        logp.append(np.nan)
        num_h_donors.append(np.nan)
        num_h_acceptors.append(np.nan)
        num_rotatable_bonds.append(np.nan)
        num_aromatic_rings.append(np.nan)
        tpsa.append(np.nan)
        num_heavy_atoms.append(np.nan)
        qed.append(np.nan)
    else:
        mols.append(mol)
        mw.append(Descriptors.MolWt(mol))
        logp.append(Descriptors.MolLogP(mol))
        num_h_donors.append(Lipinski.NumHDonors(mol))
        num_h_acceptors.append(Lipinski.NumHAcceptors(mol))
        num_rotatable_bonds.append(Descriptors.NumRotatableBonds(mol))
        num_aromatic_rings.append(Descriptors.NumAromaticRings(mol))
        tpsa.append(Descriptors.TPSA(mol))
        num_heavy_atoms.append(mol.GetNumHeavyAtoms())
        qed.append(Descriptors.qed(mol))

# Create a new dataframe with the descriptors
descriptors_df = pd.DataFrame({
    'MW': mw,
    'LogP': logp,
    'NumHDonors': num_h_donors,
    'NumHAcceptors': num_h_acceptors,
    'NumRotatableBonds': num_rotatable_bonds,
    'NumAromaticRings': num_aromatic_rings,
    'TPSA': tpsa,
    'NumHeavyAtoms': num_heavy_atoms,
    'QED': qed
})

# Add molecule ID and bioactivity class from the original dataframe
# Assuming 'molecule_chembl_id' and 'class' are columns in df_final
if 'molecule_chembl_id' in df_final.columns:
    descriptors_df['molecule_chembl_id'] = df_final['molecule_chembl_id']
    
if 'class' in df_final.columns:
    descriptors_df['bioactivity_class'] = df_final['class']

# Add pIC50 values if available
if 'pIC50' in df_final.columns:
    descriptors_df['pIC50'] = df_final['pIC50']

# Reorder columns to put identifiers first
cols = descriptors_df.columns.tolist()
identifier_cols = [col for col in cols if col in ['molecule_chembl_id', 'bioactivity_class', 'pIC50']]
descriptor_cols = [col for col in cols if col not in identifier_cols]
descriptors_df = descriptors_df[identifier_cols + descriptor_cols]

# Display the first few rows
print(descriptors_df.head())

# Save to CSV
descriptors_df.to_csv('molecular_descriptors.csv', index=False)

# Print summary statistics
print("\nDescriptor Summary Statistics:")
print(descriptors_df.describe())

# Print number of molecules processed
print(f"\nProcessed {len(descriptors_df)} molecules")
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit.Chem import AllChem, Descriptors, Lipinski
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# Assuming df_final has a column with SMILES strings
smiles_column = 'canonical_smiles'
class_column = 'class'  # Adjust if your class column has a different name

# Step 1: Calculate Lipinski's Rule of 5 descriptors for EDA
print("Calculating Ro5 descriptors...")
ro5_descriptors = []

for smiles in df_final[smiles_column]:
    mol = Chem.MolFromSmiles(smiles)
    
    if mol is None:
        ro5_descriptors.append({
            'MW': np.nan,
            'LogP': np.nan,
            'NumHDonors': np.nan,
            'NumHAcceptors': np.nan
        })
    else:
        ro5_descriptors.append({
            'MW': Descriptors.MolWt(mol),
            'LogP': Descriptors.MolLogP(mol),
            'NumHDonors': Lipinski.NumHDonors(mol),
            'NumHAcceptors': Lipinski.NumHAcceptors(mol)
        })

# Convert to DataFrame
ro5_df = pd.DataFrame(ro5_descriptors)

# Add bioactivity class
ro5_df['bioactivity_class'] = df_final[class_column]

# Step 2: Generate Morgan fingerprints for machine learning
print("Generating Morgan fingerprints...")
radius = 2  # ECFP4 equivalent
nBits = 2048  # Common choice: 1024 or 2048

# Function to generate Morgan fingerprints
def generate_morgan_fingerprints(smiles, radius=2, nBits=2048):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return np.zeros(nBits)
    
    fingerprint = AllChem.GetMorganFingerprintAsBitVect(mol, radius, nBits=nBits)
    return np.array(fingerprint)

# Generate fingerprints for all molecules
fingerprints = []
for smiles in df_final[smiles_column]:
    fingerprints.append(generate_morgan_fingerprints(smiles, radius, nBits))

# Convert to numpy array
X = np.array(fingerprints)

# Create target variable (if needed for ML)
y = df_final[class_column].map({'active': 1, 'inactive': 0, 'intermediate': 2})

# Step 3: Combine fingerprints with Ro5 descriptors (optional for ML)
# Standardize Ro5 descriptors
scaler = StandardScaler()
ro5_scaled = scaler.fit_transform(ro5_df[['MW', 'LogP', 'NumHDonors', 'NumHAcceptors']])

# Combine with fingerprints
X_combined = np.hstack((X, ro5_scaled))

# Step 4: Save the data for machine learning
print("Saving data for machine learning...")
np.save('morgan_fingerprints.npy', X)
np.save('combined_features.npy', X_combined)
np.save('target_values.npy', y.values)

# Save mapping for reference
pd.DataFrame({
    'molecule_id': df_final.get('molecule_chembl_id', df_final.index),
    'class': df_final[class_column],
    'class_numeric': y
}).to_csv('target_mapping.csv', index=False)

# Step 5: Exploratory Data Analysis of Ro5 descriptors
print("Performing EDA on Ro5 descriptors...")

# Set up the figure for visualization
plt.figure(figsize=(15, 12))

# Define colors for each class
class_colors = {'active': 'green', 'inactive': 'blue', 'intermediate': 'red'}

# Create boxplots for each Ro5 descriptor
for i, descriptor in enumerate(['MW', 'LogP', 'NumHDonors', 'NumHAcceptors']):
    plt.subplot(2, 2, i+1)
    
    sns.boxplot(x='bioactivity_class', y=descriptor, data=ro5_df, palette=class_colors)
    
    plt.title(f'{descriptor} Distribution by Bioactivity Class')
    plt.xlabel('Bioactivity Class')
    plt.ylabel(descriptor)
    
    # Add Lipinski's Rule of 5 thresholds as horizontal lines
    if descriptor == 'MW':
        plt.axhline(y=500, color='r', linestyle='--', alpha=0.5, label='Ro5 Threshold (500)')
    elif descriptor == 'LogP':
        plt.axhline(y=5, color='r', linestyle='--', alpha=0.5, label='Ro5 Threshold (5)')
    elif descriptor == 'NumHDonors':
        plt.axhline(y=5, color='r', linestyle='--', alpha=0.5, label='Ro5 Threshold (5)')
    elif descriptor == 'NumHAcceptors':
        plt.axhline(y=10, color='r', linestyle='--', alpha=0.5, label='Ro5 Threshold (10)')
    
    plt.legend()

plt.tight_layout()
plt.savefig('ro5_descriptors_by_class.pdf')
plt.show()

# Step 6: Analyze Lipinski's Rule of 5 violations
print("Analyzing Ro5 violations...")

# Function to count Ro5 violations
def count_ro5_violations(row):
    violations = 0
    if row['MW'] > 500: violations += 1
    if row['LogP'] > 5: violations += 1
    if row['NumHDonors'] > 5: violations += 1
    if row['NumHAcceptors'] > 10: violations += 1
    return violations

# Count violations for each molecule
ro5_df['Ro5_Violations'] = ro5_df.apply(count_ro5_violations, axis=1)

# Create a bar plot of violation counts by class
plt.figure(figsize=(10, 6))
sns.countplot(x='Ro5_Violations', hue='bioactivity_class', data=ro5_df, palette=class_colors)
plt.title('Lipinski\'s Rule of 5 Violations by Bioactivity Class')
plt.xlabel('Number of Violations')
plt.ylabel('Count')
plt.legend(title='Bioactivity Class')
plt.savefig('ro5_violations_by_class.pdf')
plt.show()

# Print summary of fingerprints and features
print(f"\nMorgan fingerprints shape: {X.shape}")
print(f"Combined features shape: {X_combined.shape}")
print(f"Target values shape: {y.shape}")
print("\nFeature generation complete!")
```
